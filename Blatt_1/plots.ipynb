```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exercise 1

In the following cell you'll find the running time of our Monte Carlo implementation based on the number of generated points. The O(n) running time complexity is clear. The spikes are probably due to other processes running in the background. 

```python
# Path to the CSV file
csv_file = 'running_time.csv'

# Load the CSV file using numpy
data = np.loadtxt(csv_file, delimiter=',', skiprows=1)

# Extract the columns from the data
NumberOfPoints_1 = data[:, 0]
Time_1 = data[:, 1]

# Plot the data
plt.plot(NumberOfPoints_1, Time_1, color='b')
plt.xlabel('Number of generated points')
plt.ylabel('Time taken [s]')
plt.title('Exercise 1: MonteCarlo Pi')
plt.show()
```

## Exercise 2

Here the running time is again plotted. A O(n) running time is again to be seen, with smaller constant factors, due to the parallelization

```python
# Path to the CSV file
csv_file = 'running_time2.csv'

# Load the CSV file using numpy
data = np.loadtxt(csv_file, delimiter=',', skiprows=1)

# Extract the columns from the data
NumberOfPoints_2 = data[:, 0]
Time_2 = data[:, 1]

# Plot the data
plt.plot(NumberOfPoints_2, Time_2, color='g')
plt.xlabel('Number of generated points')
plt.ylabel('Time taken [s]')
plt.title('Exercise 2: MonteCarlo Pi with OpenMP')
plt.show()
```

### Comparing running times

In the following plots a direct comparison is show. In Figure 3 it is clear that the implementation using OpenMP is clearly faster, while in the Figure 4 we see the absolute speedup depending on the size of the input. We see a speedup that trends towards 4, which is to be expected since it was performed on a machine with 4 threads.

```python
# Plot the data
plt.plot(NumberOfPoints_1, Time_1, label='No parallelization', color='b')
plt.plot(NumberOfPoints_2, Time_2, label='OpenMP', color='g')
plt.legend()
plt.xlabel('Number of generated points')
plt.ylabel('Time taken [s]')
plt.title('Figure 3: MonteCarlo Pi with OpenMP')
plt.show()

plt.plot(NumberOfPoints_1, Time_1/Time_2)
plt.xlabel('Number of generated points')
plt.ylabel('Absolute speedup')
plt.title('Figure 4: Absolute speedup')
plt.show()
```

### Exercise 3

In the following the results of the implementation with MPI (here OpenMPI) are presented. A running time around O(n) is appreciated even tho many lag spikes are present.

```python
# Path to the CSV file
csv_file = 'running_time3.csv'

# Load the CSV file using numpy
data = np.loadtxt(csv_file, delimiter=',', skiprows=1)

# Extract the columns from the data
NumberOfPoints_3 = data[:, 0]
Time_3 = data[:, 1]

# Plot the data
plt.plot(NumberOfPoints_3, Time_3, color='r')
plt.xlabel('Number of generated points')
plt.ylabel('Time taken [s]')
plt.title('Exercise 3: MonteCarlo Pi with MPI')
plt.show()
```

### Final comparison

It is clear that the sequential approach is the slowest among all of them. The implementation with MPI seems to be faulty. probably the saving of the data was not done in the correct way. It could also be that the message passing is not done efficiently enough.

```python
# Plot the data
plt.plot(NumberOfPoints_1, Time_1, label='No parallelization', color='b')
plt.plot(NumberOfPoints_2, Time_2, label='OpenMP', color='g')
plt.plot(NumberOfPoints_3, Time_3, label='MPI', color='r')
plt.legend()
plt.xlabel('Number of generated points')
plt.ylabel('Time taken [s]')
plt.title('Figure 5: Comparison of sequential, OpenMP and MPI')
plt.show()

plt.plot(NumberOfPoints_1, Time_1/Time_3)
plt.xlabel('Number of generated points')
plt.ylabel('Absolute speedup')
plt.title('Figure 6: Absolute speedup MPI')
plt.show()
```
